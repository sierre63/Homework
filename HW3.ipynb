```python
import os
from pymystem3 import Mystem #Импортируем Mystem
import re #импортируем регулярные выражения
def my_function(x,y):
    with open(x, 'r', encoding = 'utf-8') as z: #открываем файл с анекдотами для чтения
        text = z.read()
    v = Mystem()
    lemmas = v.lemmatize(text) #лемматизируем файл
    v3 = ' '.join(lemmas) #объединяем в строку
    v3 = re.sub(r'[^\w\s]', '', v3) #убираем знаки препинания
    
    
    with open(y, 'a+', encoding = 'utf-8') as b:   #Создаем новый файл и открываем для чтения
        b.write(v3) 

my_function('anecdotes.txt' , 'lemmas.txt') #Вызываем функцию
```

```python
with open('lemmas.txt', 'r', encoding = 'utf-8') as b: #открываем новый файл с леммами для чтения
    lemmas2 = b.read()
    with open("rus_stopwords.txt", "r", encoding = "utf-8") as e: #открываем файл стоп-словами для чтения
        stopariki = e.read()
    
        lemmas3 = lemmas2.split() #разбиваем по словам 1й файл
        count_lemmas = len(lemmas3) #считаем количество слов в нем
        
        stopariki2 = stopariki.split() #разбиваем по словам файл со стоп-словами

        cleanlist = [q for q in lemmas3 if q not in stopariki2]  #убираем стоп-слова

        from collections import Counter #импортируем Counter
        end = Counter(cleanlist) #считаем частотность слов
        ipm = {w: end[w]/count_lemmas*1000000 for w in end} #считаем относительную частотность (ipm)
        ipm = sorted(ipm.items(), key=lambda item: item[1], reverse = True) #сортируем словарь 

print(ipm[0:20]) #частотность (ipm) - 20 слов
print(end.most_common(20)) #абсолютная частотность - 20 слов
```

```python
with open('anecdotes.txt', 'r', encoding = 'utf-8') as c: #открываем исходый файл для чтения
    morfologia = c.read()
    
    stem = Mystem()
    analyze = stem.analyze(morfologia) #анализируем файл на морфологию
    
    morf = re.findall('[A-Z]+', str(analyze)) #ищем все заглавные латинские значения - части речи + полный морфологический анализ текста
    itog = {} #создаем словарь
    
    for slovo in analyze: #для каждого слова в анализе ищем часть речи и создаем словарь
        try:
            gram = slovo['analysis'][0]['gr']
            chast_rechi = re.search('[A-Z]+', gram).group()
            forma = slovo['text']
            itog[forma] = chast_rechi
            
        except (KeyError, IndexError): #исключаем ошибки, потому что они появляются и я не знаю, что делать
            pass
        
    print(itog)
```

```python
chastota = Counter(morf)
print(chastota) #считаем абсолютную частоту
```

```python
kolvo_slov = len(morfologia.split())
otn_chastota = {w: chastota[w]/kolvo_slov for w in chastota} #считаем относительную частотность
otn_chastota = sorted(otn_chastota.items(), key=lambda item: item[1], reverse = True)
print(otn_chastota)
```
