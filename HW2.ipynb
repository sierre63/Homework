```python
# 1 задание
poem='В новом море технологий важно давать людям шанс. В компании Amazon внедрена культура: если у руководителя нижнего звена есть идея и он хочет ее опробовать, ему нужно убедить своего начальника, а тому - своего и так далее.'
x=poem.split() #делим по словам
print(x)
bigrams = []
for e in range(0, len(x)):
    if e != len(x) - 1: 
        bigrams.append((x[e], x[e + 1])) #добавляем биграммы
print(bigrams)
```

```python
# 3 задание
text="""Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!"""

print(len(text)) #считаем количество символов

from string import punctuation 
text2 = text.split()
text2=[w.strip(punctuation) for w in text.split()] #убираем знаки пунктуации
print(text2)

text3=" ".join(text2) #объединяем слова без пунктуации
print(text3)

koshka = []
for d in text3.split(): #делим текст по словам
    if d.istitle():
        koshka.append(d.lower()) 
    else:
        koshka.append(d) #все слова с заглавными буквами переводим в строчные
print(koshka)

koshka1=text.split() #делим по словам
print(len(koshka1)) #выводим количество слов

koshka2=set(koshka1) 
print(len(koshka2)) #считаем уникальные слова и их количество

print(koshka1[-10:]) #выводим последние 10 слов

koshka3='\n'.join(text2) #выводим текст без пунктуации построчно
print(koshka3)
```
